```python
# Вычисление отклонений курсов от средних значений
data['rate_1_deviation_1m'] = data['rate1'] - data['w1m_0mos_avg_rate1']
data['rate_2_deviation_1m'] = data['rate2'] - data['w1m_0mos_avg_rate2']

data['rate_1_deviation_5m'] = data['rate1'] - data['w1m_5mos_avg_rate1']
data['rate_2_deviation_5m'] = data['rate2'] - data['w1m_5mos_avg_rate2']

data['rate_1_deviation_1h'] = data['rate1'] - data['w1h_1mos_corr']
data['rate_2_deviation_1h'] = data['rate2'] - data['w1h_1mos_corr']

# Добавление столбца 'profit' для анализа успешности сделок
data['profit'] = data['profit_percent'].apply(lambda x: 1 if x > 0 else 0)

# Исторические значения курсов
historical_max_rate_1 = data['rate1'].max()
historical_min_rate_1 = data['rate1'].min()

historical_max_rate_2 = data['rate2'].max()
historical_min_rate_2 = data['rate2'].min()

# Проверка текущих значений курсов rate1 и rate2 на экстремальные значения
data['open_at_extreme'] = ((data['rate1'] == historical_max_rate_1) | (data['rate1'] == historical_min_rate_1) |
                           (data['rate2'] == historical_max_rate_2) | (data['rate2'] == historical_min_rate_2))

# Определение резких импульсов (например, изменение больше чем 2 стандартных отклонения)
mean_change_1m = data['rate_1_deviation_1m'].mean()
std_dev_change_1m = data['rate_1_deviation_1m'].std()
data['sharp_impulse_1m'] = data['rate_1_deviation_1m'].apply(lambda x: 1 if abs(x) > 2 * std_dev_change_1m else 0)

mean_change_5m = data['rate_1_deviation_5m'].mean()
std_dev_change_5m = data['rate_1_deviation_5m'].std()
data['sharp_impulse_5m'] = data['rate_1_deviation_5m'].apply(lambda x: 1 if abs(x) > 2 * std_dev_change_5m else 0)

mean_change_1h = data['rate_1_deviation_1h'].mean()
std_dev_change_1h = data['rate_1_deviation_1h'].std()
data['sharp_impulse_1h'] = data['rate_1_deviation_1h'].apply(lambda x: 1 if abs(x) > 2 * std_dev_change_1h else 0)

# Открытие сделок при наличии резких импульсов или экстремальных значений курсов
data['open_trade'] = data['sharp_impulse_1m'] | data['sharp_impulse_5m'] | data['sharp_impulse_1h'] | data['open_at_extreme']

# Успешность сделок с импульсами и экстремальными значениями
success_rate_with_conditions = data[data['open_trade']]['profit'].mean()

# Успешность сделок без импульсов и экстремальных значений
data['no_impulse'] = ~(data['open_trade'])
success_rate_without_conditions = data[data['no_impulse']]['profit'].mean()

print(f"Успешность сделок с учетом резких импульсов и экстремальных значений курсов: {success_rate_with_conditions * 100:.2f}%")
print(f"Успешность сделок без резких импульсов и экстремальных значений курсов: {success_rate_without_conditions * 100:.2f}%")

```
